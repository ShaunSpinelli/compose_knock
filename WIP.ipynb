```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import os
import pprint
import pickle
import random
```

```python
url = "http://www.shortjokes.com.au/jokes-Knock-Knock-jokes-page-"
```

```python
jokes=[]
for i in range(0,368,20):
    r  = requests.get(f'{url}{i}.html')
    soup=BeautifulSoup(r.text,'lxml')
    jks = soup.find_all('span', {'class':'joke'}) 
    jokes += [i.findChild().findChild().contents[0].replace('KnockWhos', 'Knock. Who\'s')\
              .replace('Knock Knock','Knock, Knock').replace(' !','! ').replace(' ?','? ')\
              .lower().strip() for i in jks]    
```

```python
jokes2 = []
x=""
for i in ds.split('\n'):
    if i!='':
        x+=str(i)
    else:
        if x!='':
            jokes2.append(x.lower())
        x=""
```

```python
results=[]
for c in range(100):
    tl=[]
    for i in range(1000):
        if random.random()>=.2:
            tl.append(i)
    results.append(len(tl))
results=np.array(results)
```

```python
results.mean()
```

```python
def write_file(title, body):
    file = open(f'{title}.txt','w')
    file.write(body)
    file.close()
    pass
```

```python
p=.2
```

```python
for j in data:
    if random.random()>=p: #member of train set
        write_file(f'train/train_{data.index(j)}', j)
    else: #member of validation set
        write_file(f'test/test_{data.index(j)}',j)
```
